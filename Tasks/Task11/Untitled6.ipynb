```python
import pandas as pd

# Load your dataset
df = pd.read_csv("PC price prediction.csv")  # Replace with actual name

# Fix column names (already done before)
df.rename(columns={
    'ram_(sistem_belle?i)_tipi': 'ram_type',
    '??lemci_?ekirdek_say?s?': 'cpu_cores',
    '??lemci_nesli': 'cpu_generation',
    '??lemci_modeli': 'cpu_model',
    '??z?n?rl?k': 'resolution',
    'kullan?m_amac?': 'usage_purpose'
}, inplace=True)

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Clean numeric object columns
def extract_number(s):
    return pd.to_numeric(str(s).split()[0].replace(',', '.').replace('"', ''), errors='coerce')

df['ram'] = df['ram'].apply(extract_number)
df['ssd_capacity'] = df['ssd_capacity'].apply(extract_number)
df['graphics_card_capacity'] = df['graphics_card_capacity'].apply(extract_number)
df['graphics_card_memory'] = df['graphics_card_memory'].apply(extract_number)
df['base_cpu_speed_(ghz)'] = df['base_cpu_speed_(ghz)'].apply(extract_number)
df['screen_refresh_rate'] = df['screen_refresh_rate'].apply(extract_number)
df['processor_frequency'] = df['processor_frequency'].apply(extract_number)
df['device_weight'] = df['device_weight'].str.extract(r'(\d+\.?\d*)').astype(float)
df['screen_size'] = df['screen_size'].str.extract(r'(\d+\.?\d*)').astype(float)
df['max_expandable_memory'] = df['max_expandable_memory'].apply(extract_number)

# Final dtype check
print(df.dtypes)

# Preview cleaned data
print(df.head())

# Final statistics
print(df.describe())
```

```python
df.to_csv("cleaned_pc_price_prediction.csv", index=False)
```
